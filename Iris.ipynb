# **Using sklearn.datasets iris flower dataset to train a model using logistic regression.**In iris dataset there are 150 samples containing following features,

Sepal Length,
Sepal Width,
Petal Length,
Petal Width

Using above 4 features let's clasify a flower in one of the three categories,
Setosa,
Versicolour,
Virginica

```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
```

```python
iris.data[0]
```

```python
dir(iris)
```

Creating and training a logistic regression model

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(iris.data,iris.target)
```

```python
model.fit(x_train,y_train)
```

```python
model.score(x_train,y_train)
```

```python
model.predict(iris.data[0:5])
```

```python
from sklearn.metrics import confusion_matrix
y_predicted = model.predict(x_test)
cm = confusion_matrix(y_test, y_predicted)

import seaborn as sn
plt.figure(figsize = (10,7))
sn.heatmap(cm, annot=True)
plt.xlabel('Predicted')
plt.ylabel('Truth')
```
